```python
import pandas as pd
import numpy as np
```

```python
data = {"a": [4, 5, 6, 6],"b": [7, 8, 9, 9],"c": [10, 11, 12, 12]}
# df = pd.DataFrame(data,index=["a","b","c","d"])
df = pd.DataFrame(data,index=list("abcd"))
df
```

#### 1) != : Not equal to

```python
# b 컬럼의 값 중 7과 같지 않은 것

df["b"] != 7
```

```python
# b 컬럼의 값 중 7과 같지 않은 것을 찾아서 DataFrame 형태로

df[df["b"] != 7]
```

```python
# a, b 컬럼에서 b 컬럼이 가지고 있는 값이 7이 아닌 것들만 보고 싶다면?

# df[df["b"] != 7][["a","b"]]
df.loc[df["b"] != 7, ["a","b"]]
```

#### 2) isin() : 지정된 값이 들어 있느냐?

```python
df.a.isin?
```

```python
# 컬럼의 값이 한글이나 특수문자 인 경우에는 대괄호 쓰는게 좋음
df.a.isin([5])
```

```python
# df[df.a.isin([5])]

df[df.a==5]
```

#### 3) isnull()

```python
# np.nan 아직 값이 입력되지 않음
# 데이터에 np.nan가 있으면 정수도 실수 형태로 변환
df = pd.DataFrame(
    {
        "a":[4, 5, 6, 6, np.nan],
        "b":[7, 8, np.nan, 9, 9],
        "c":[10, 11, 12, np.nan, 12]
    },
    index=[1,2,3,4,5]
)
df
```

```python
# 전체 널값 확인

df.isnull()
```

```python
# a 컬럼에 null 값이 있는가?

df["a"].isnull()
```

```python
# 널 개수

df.isnull().sum()
```

#### 4) isna()

```python
# isnull()과 비슷
df.isna()
```

```python
df.isna().sum()
```

```python
# 에러 남
# df[df.b == 7] and [df.a == 5]
# 파이썬 문법에서 맞는 것들 일부는 pandas에서는 안됨
```

```python
# b 컬럼이 7이고 a 컬럼이 5인 행 찾기
# df[df.b == 7] and [df.a == 5]
df[(df.b == 7) & (df.a == 5)]
```

```python
# b 컬럼이 7이거나 a 컬럼이 5인 행 찾기
df[(df.b == 7) | (df.a == 5)]
```
