```python
from konlpy.tag import Okt
okt = Okt()
```

```python
# 명사 추출

# okt.nouns(df_news['message'][0])

# okt.nouns(df_news['message']) 는 안됨
nouns = okt.nouns(" ".join(df_news['message']))
```

```python
from collections import Counter
```

```python
count = Counter(nouns)
# {'스파크': 3, ... 의 딕셔너리 형식
count
```

```python
for k, v in count.items():
  print(k, v)
```

```python
df_word = pd.DataFrame(
    {
        "word": count.keys(),
        "count" : count.values()
    }
)
```

```python
df_word
```

```python
from time import asctime
```

```python
# df_word[df_word['word'].str.len() > 1]
# df_word[df_word['word'].apply(lambda x:len(x) > 1)]
df_word[df_word['word'].apply(lambda x:len(x) > 1)].sort_values('count', ascending=False)[:50]
```

```python
# 등장빈도 높은 단어 그래프 만들기
df_word[df_word['word'].apply(lambda x:len(x) > 1)].sort_values('count', ascending=False)[:50].plot.bar(x='word', y='count', figsize=(20,5), rot=40)
```
