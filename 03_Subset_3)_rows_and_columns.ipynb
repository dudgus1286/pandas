### 인덱싱 적용 방법

|적용방법|구문|결과타입|
|:---|:---:|:---:|
|열선택|df[col]|Series|
|라벨로 행 선택|df.loc[label]|Series|
|정수 위치로 행 선택|df.iloc[loc]|Series|
|행 슬라이스|df[5:10]|DataFrame|
|불리언 벡터로 행 선택|df[bool_vec]|DataFrame|

````python
import pandas as pd
import seaborn as sns
import numpy as np
````

````python
friend_dict_list = [
    {
        "name": "John",
        "age": 25,
        "job": "student",
    },
    {
        "name": "Nate",
        "age": 34,
        "job": "teacher",
    },
    {
        "name": "Jenny",
        "age": 30,
        "job": "developer",
    },
]

friend_df = pd.DataFrame(friend_dict_list)
friend_df
````

### rows, columns 기준

#### loc[]
- 5 또는 a 와 같은 단일 라벨
- ['a','b','c']와 같은 리스트나 라벨의 배열
- a:f 와 같이 라벨이 있는 처음과 끝의 범위를 포함하는 슬라이스 객체
- [True, False, True]와 같이 슬라이싱할 축과 같은 길이의 불리언 배열
- 호출하는 시리즈나 데이터 프레임을 인수로 가지는 호출 함수

#### iloc[]
- 정수
- [4,3,0] 과 같은 리스트나 정수들의 배열
- 1:7 과 같은 정숫값을 가지는 슬라이스 객체
- 불리언 배열
- 호출하는 시리즈나 데이터프레임을 인수로 가지는 호출 함수

````python
# 인덱스 라벨 : 0
# friend_df.loc[0]

# 정수(기본으로 부여되는 숫자값) 으로 조회
# 시리즈로 반환
friend_df.iloc[0]
````

````python
# 슬라이싱 : 데이터프레임으로 반환
friend_df.iloc[0:1]
````

````python
friend_df.iloc[1:3]
````

````python
# 기본 데이터 셋 가져오기
iris_df=sns.load_dataset("iris")
iris_df.head()
````

><b>df.iloc[ ] : 행, 열 모두 position 값을 이용(integer 값)</b>

````python
iris_df.iloc[10:20]
````

````python
iris_df.iloc[1:3,1:3]
````

````python
# 행 index는 모두 포함, 열 index는 1, 3번 컬럼만 포함

iris_df.iloc[:,[1, 3]]
````

````python
iris_df.iloc[:, [1,2,4]]
````

><b>df.loc[ ] : 행, 열의 label 값을 이용하거나, boolean 이용 / 마지막 값 포함</b>

````python
iris_df.loc[1]
````

````python
iris_df.loc[11:21]
````

````python
iris_df.loc[1, "sepal_width":"petal_width"]
iris_df.loc[1][["sepal_width", "petal_length", "petal_width"]]
````

````python
iris_df.loc[20:30, "sepal_width":"petal_width"]
iris_df.loc[20:30][["sepal_width", "petal_length", "petal_width"]]
````

````python
# iris_df.loc[:, "sepal_width":"petal_width"]
iris_df.loc[:][["sepal_width", "petal_length", "petal_width"]]
````

````python
iris_df.loc[[1,3], ["sepal_width","petal_width"]]
````

````python
# sepal_width > 3.0
iris_df.loc[iris_df["sepal_width"]> 3.0]
````

````python
# sepal_width > 3.0 , sepal_width,	petal_width 만 조회
iris_df.loc[iris_df["sepal_width"]> 3.0, ["sepal_width","petal_width"]]
````

#### 3) filter(items=None, like=None, regex=None, axis=None)
```
행 또는 열의 부분 집합 지정
```

````python
# _가 들어있는 컬럼 가져오기

iris_df.filter(regex="_")
````

````python
# length 로 끝나는 컬럼 가져오기

# ~로 끝나는 : $
# ~로 시작하는 : ^
iris_df.filter(regex="length$")
````

````python
# Sepal 로 시작하는 컬럼 가져오기

iris_df.filter(regex="^sepal")
````

````python
# se로 시작하는 컬럼

iris_df.filter(regex="^se")
````

````python
# data_dict_list = {
#     "one" : [1,4],
#     "two" : [2,5],
#     "three" : [3,6]
# }

# data = pd.DataFrame(data_dict_list, index=["mouse", "rabbit"])
data = pd.DataFrame(data=[[1,2,3],[4,5,6]], index=["mouse", "rabbit"], columns=["one", "two", "three"])
data
````

````python
data.filter(items=["one","three"])
````

````python
# axis = 검색할 열, 행 방향 변경
data.filter(regex="e$", axis=1)
````

````python
# like = 포함 여부
data.filter(like="bbi", axis=0)
````
