### 3) sort_values()

```python
# mpg 컬럼을 기준으로 정렬하기

df.sort_values('mpg')
```

```python
# mpg 컬럼을 기준으로 정렬하기 - 내림차순

df.sort_values('mpg',ascending=False)
```

### 4) rename

```python
df.rename?
```

```python
# model_year 컬럼명 변경
# columns = {"바꿀컬럼명":"새컬럼명"}

df.rename(columns = {'model_year':'year'})
```

### 6) reset_index() : index 를 컬럼으로 만들기(앞에서 사용함)

```python
# indext였던 걸 컬럼으로 변경하는 개념이기 때문에 기존에는 실제 존재하는 컬럼을 인덱스로 만들고 다시 컬럼으로 만드는 것이었기 때문에
# df.reset_index("년도") 이렇게 사용한 것

df.head()
df.set_index("cylinders")
df.reset_index()
# 기존 인덱스(0~)도 컬럼으로 나옴
```

### 5) sort_index() : index 를 기준으로 정렬하기

```python
# sort_values("컬럼명") : 해당 컬럼을 기준으로 정렬

df.sort_index(ascending=False)   # 위에서 만든 index 를 기준으로 정렬
```

### 7) drop(columns = ['','','']) : 괄호 안에 있는 컬럼 삭제 / drop(["",""],axis=1)

```python
# drop_duplicates() / duplicated() - True/False 값에 기반해서 중복 제거

df.drop(['mpg','model_year'],axis=1)
```

```python
df.drop(columns = ['mpg','model_year'])  # df.drop(['mpg','model_year'],axis=1)
```

### [실습2]

```python
df = pd.DataFrame({"년도":[2019,2020,2021],"미국":[2.1,2.2,2.3], "한국":[0.4,0.5,0.45], "중국":[10,13,15]})
df
```

### 1) 인덱스 생성

```python
# 년도를 index로 생성하기

df = df.set_index("년도")
df
```

### 2) 인덱스 제거

```python
# 인덱스를 컬럼으로 변경하기

df = df.reset_index('년도')
df
```

### 3) 컬럼명 변경(중국==> 홍콩)

```python
df.rename(columns={'중국':'홍콩'},inplace=True)
df
```
