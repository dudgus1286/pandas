### [실습1]

```python
df = sns.load_dataset("mpg")
df.head()
```

```python
df.groupby("origin").size()
```

```python
# value_counts("컬럼명")
df.value_counts("origin")
```

```python
# origin 별로 평균값을 보고 싶다면?

df.groupby("origin").mean(numeric_only = True)
```

### transform()
- transform('집계함수') : 연산의 결과를 현재 데이터프레임의 인덱스에 맞춰서 돌려줌

### 실습

```python
df = pd.DataFrame(
    {
        "School" : ["Yeonhi", "Yeonhi", "Sungsan", "Sungsan", "Sungsan"],
        "Name" : ["Haena", "Gisu", "Una", "Naeun", "Ziho"],
        "Math_s" : [92,70,88,92,70]
    }
)
df
```

```python
# school을 그룹으로 하여 수학 성적의 평균 구하기
df2 = df.groupby("School")["Math_s"].mean().reset_index().rename(columns={"Math_s":"Avg_s"})
```

```python
df3 = pd.merge(df, df2)
df3
```

- transform() 을 사용시

```python
df2_1 =df.groupby("School")["Math_s"].transform("mean")
df["Avg_s"] = df2_1
df
```

- apply() : apply(적용할함수)

```python
# >= 90 : A, >= 80 : B, >= 70 : C, 그 외 F

def grade(num):
  if num >= 90:
    return "A"
  elif num >= 80:
    return "B"
  elif num >= 70:
    return "C"
  else: return "F"
```

```python
# 함수 동작 확인

grade(95)
```

```python
df['grade'] = df['Math_s'].apply(grade)
# map(grade)도 가능, applymap는 불가능
df
```

```python
df3['grade'] = df3["Math_s"].map(grade)
df3
```

```python
df3.sort_values("Math_s", ascending=False)
```

```python
def sort_math(df, column="Math_s", n=3):

  return df.sort_values(column, ascending=False)[:n]
```

```python
df3.groupby('School').apply(sort_math,n=2)
```
