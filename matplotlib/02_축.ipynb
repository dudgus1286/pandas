```python
# import 구문
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
```

```python
# 설치된 글꼴 알아보기

all_fonts_I_can_use = fm.findSystemFonts(fontpaths=None, fontext='ttf')
all_fonts_I_can_use
```

```python
# matplotlib 에서 해당 폰트를 사용할 수 있도록 등록

for f_path in all_fonts_I_can_use:
  fm.fontManager.addfont(f_path)

# matplotlib 기본 글꼴 설정
plt.rc("font", family="NanumBarunGothic")
```

```python
x = [1,2,3]
y = [2,4,8]

plt.plot(x, y);
```

### 타이틀 설정
- https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.title.html#matplotlib.pyplot.title

```python
# fontdict 딕셔너리 형태로 폰트 관련 설정 지정, 설정 안 하면 기본값으로 출력
# rcParams['axes.titlesize'] : 사용자가 별도로 설정한 값을 가져오도록 함
plt.title("꺽은선 그래프", fontdict={
    "fontsize" : 16
});
plt.plot(x,y);
```

### 축설정
- https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xlabel.html
- https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.ylabel.html

#### 1. 이름 설정
><b>1) xlabel(), ylabel()</b>

```python
plt.title("꺽은선 그래프", fontdict={"fontsize": 16});
plt.plot(x,y)
plt.xlabel("X 축", color="red", loc="right")
plt.ylabel("Y 축", color ="#00aa00", loc="top");
```

><b>2) 이름 및 다양한 설정</b>

matplotlib.pyplot.xlabel(xlabel, fontdict=None, labelpad=None, *, loc=None, **kwargs)

- fontdict : font 설정을 dict 구조로 설정
- labelpad : default 4.0 설정됨(label과 차트 사이의 간격)
- **kwargs
    - color : rgb설정, 색상명 가능
- loc : 축 이름 위치 설정('left', 'center', 'right'),  y 축 (top, bottom, center)

```python
# def xlabel(xlabel, fontdict=None, labelpad=None, *, loc=None, **kwargs)
# **kwargs : 키 = 값 형태, color="red", fontsize=20 등으로 부여한 속성을 다룸 (옵션)

# xlabel("x축 이름"(필수), {}, labelpad=1, loc="", )
```

```python
fontdict1 = {"family": "serif", "color": "blue", "weight":"bold", "size": 14}
fontdict2 = {"family": "serif", "color": "deeppink", "weight":"bold", "size": 15}

plt.title("꺽은선 그래프", fontdict={"fontsize": 16});
plt.plot(x,y)
plt.xlabel("X", fontdict=fontdict1, loc="right", labelpad = 15)
plt.ylabel("Y", fontdict=fontdict2, loc="top", labelpad = 20);
```

#### 2. 축 눈금 설정 - xticks(), yticks()

```python
plt.plot(x,y)

xticks = [1,2,3]
yticks = [3,6,9,12]

plt.xticks(xticks, fontsize=15)
plt.yticks(yticks, fontsize=20);
```
