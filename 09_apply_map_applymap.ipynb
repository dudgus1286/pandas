## apply

- https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.apply.html
- apply 라는 함수를 적용할 때 특정 컬럼에 적용하는 형태 이거나, 전체 df에 적용하는 형태 가능
- ex) df.apply(함수), df['컬럼명'].apply(함수)
- df.apply 일 때 중요한 부분은 방향임

```
DataFrame.apply(func, axis=0, raw=False, result_type=None, args=(), **kwargs)

axis = 0 이라는 소리는 하나의 컬럼에 속하는 모든 행에 적용된다는 의미

axis = 1 이라는 소리는 하나의 행에 있는 모든 컬럼이 같이 적용 된다는 의미

```




````python
import pandas as pd
import numpy as np
import seaborn as sns
````

### [실습1]

````python
df = pd.DataFrame({
    "영어":[60,70],
    "수학":[100, 50]
}, index=["Dave", "David"])
df
````

````python
# 함수 정의

def func(df):
  print(type(df))
  print(df.index)
  print(df.values)
  return df
````

````python
# 결국 열에 함수가 적용

df.apply(func)
# == func(df)
````

````python
# 행기준을 적용시키고 싶을 때

df.apply(func, axis=1)
````

````python
# apply 를 통해 데이터 값 수정

def func2(df):
  df["영어"] = 80
  return df
````

````python
df.apply(func2, axis=1)
# axis 설정 안 하면 영어라는 행이 추가됨
````

### [실습3]

````python
df = pd.DataFrame(["2005-09-28", "2007-10-05", "2012-12-20"], columns=["yyyy-mm-dd"])
df
````

#### 1) 년-월-일 순으로 되어 있는 데이터를 받아서 년도만 잘라내서 리턴

````python
# df["yyyy-mm-dd"].str.split("-")

# "2005-09-28".split("-")[0]

def extract_year(date):
  return date.split("-")[0]
````

````python
df["yyyy-mm-dd"].apply(extract_year)
````

````python
df["year"] = df["yyyy-mm-dd"].apply(extract_year)
df
````

#### 2) 적용할 함수와 다른 인자를 넘길 수 있는가?

````python
def get_age(year, current_year):
  return current_year - int(year)
````

````python
df["age"]=df["year"].apply(get_age, current_year= 2024)
df
````

````python
# introduce 작성

def get_introduce(age, prefix, suffix):
  return prefix + str(age) + suffix
````

````python
# age 필드의 값을 가져와서 get_introduce 적용

df['introduce']=df["age"].apply(get_introduce, prefix = "나는 ", suffix = " 입니다.")
df
````

#### 3) 여러개의 컬럼을 가져와서 apply() 적용하기

````python
def get_introduce2(df):
  return "나는 " + str(df['year']) + " 년에 태어났고, 나의 나이는 " + str(df['age']) + " 입니다."
````

````python
df["introduce2"] = df.apply(get_introduce2, axis=1)
df
````

### lambda

## map

- https://pandas.pydata.org/docs/reference/api/pandas.Series.map.html
- apply() 와 동일하나, 데이터프레임에는 적용 못함(<b style='color:red'>시리즈</b>만 가능)
```
Series.map(arg, na_action=None)
arg : function, dict, series 가능
```

### [실습1]

````python
s = pd.Series(["cat", "dog", np.nan, "rabbit"])
s
````

````python
# cat ==> kitten / dog ==> puppy
# s.loc[0] = "kitten"
s.map({"cat":"kitten", "dog":"puppy"})
````

- map 은 dict or Series 를 받아 들임
- 값을 찾지 못하면 dict 의 경우는 NaN 으로 변환, default value 가 있다면 그 값으로

### [실습2]

````python
df = pd.DataFrame({"yyyy-mm-dd": ["2005-09-28", "2007-10-05", "2012-12-20"]})
df
````

````python
# 년도 추출

df['year'] = df['yyyy-mm-dd'].map(extract_year)
df
````

### [실습3]

````python
df = pd.DataFrame({"age":[20,30,30], "job":["student", "developer", "teacher"]})
df
````

````python
# job_id 컬럼 생성
# student : 1 , developer : 2, teacher : 3 번호 부여

df["job_id"] = df["job"].map({"student":1, "developer":2, "teacher":3})
df
````

## applymap

- https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.applymap.html
- 모든 값에 적용
- 시리즈 형태에다가는 적용 불가(<b style='color:red'>데이터프레임</b>만 가능)
- df['컬럼명'].applymap() => X 안됨

```
DataFrame.applymap(func, na_action=None, **kwargs)[source]
```

### [실습1]

````python
df = pd.DataFrame(
    [
        [1,2.12],
        [3.356, 4.567]
    ]
)
df
````

````python
df.dtypes
````

````python
x = 10.1

print(type(x))

# str() : 데이터타입을 문자열로 변경
print(type(str(x)))
````

````python
# float64 => object 로 변경

df = df.applymap(lambda x : str(x))
print(df.dtypes)

df
````

````python
# df[0].applymap(lambda x : str(x)) < - 에러 발생

# 특정 컬럼만 함수 적용해야 한다면 둘 중에 선택
df[0].apply(lambda x : str(x))
df[1].map(lambda x : str(x))
````

````python
df.dtypes

df.applymap(lambda x : len(x))
````
