```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
```

```python
for fpath in fm.findSystemFonts(fontpaths=None, fontext='ttf'):
  fm.fontManager.addfont(fpath)

plt.rc("font",family='NanumBarunGothic')
```

```python
x = [1,2,3]
y = [2,4,8]
```

```python
plt.plot(x,y);
```

```python
days=[1,2,3]
list1=[2,4,8]
list2=[5,1,3]
list3=[1,2,5]

plt.plot(days,list1)
plt.plot(days,list2)
plt.plot(days,list3);
```

```python
plt.plot(days,list1, label="az")
plt.plot(days,list2,"o--", label="pfizer")
plt.plot(days,list3, "s-.", label="moderna")
plt.legend();
```

```python
plt.plot(days,list1)
plt.plot(days,list2,marker="o", ls="--")
plt.plot(days,list3, marker="s", ls="-.")
plt.legend(["az","pfizer", "moderna"],ncols=3);
```
