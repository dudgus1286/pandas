## Summarize Data

```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = sns.load_dataset("iris")
df
```

#### 1) value_counts() :  unique 값들의 행 수 카운트

```python
df["species"].value_counts()
```

```python
df["sepal_length"].value_counts()
```

```python
# 시리즈 결과를 데이터프레임으로 생성
species_df = pd.DataFrame(df["species"].value_counts())
species_df
```

#### 2) len() :  데이터프레임 안의 행의 개수

```python
len(df)
```

#### 3) shape :  행, 열의 개수를 튜플로

```python
df.shape
```

```python
# shape에서 행 가져오기
df.shape[0]
```

#### 4) unique() / nunique

```python
df["species"].unique()
```

```python
df["species"].nunique()
```

#### 5) describe() :  기술 통계 정보(분산,최대,최소,count,표준편차,quantile)

```python
# object 타입 제외, 기술 통계 정보
df.describe()
```

```python
df.describe(include="all")
```

```python
df.describe(include=object)
```

```python
df.describe(include=np.number)
```

#### 6) info() :  데이터 프레임 전체 정보

```python
df.info()
```

#### 7) sum()

```python
# 합계 구하기
df.sum()
```

```python
df["sepal_length"].sum()
```

#### 8) count()

```python
df.count()
```

```python
df["sepal_length"].count()
```

#### 9) mean() : 평균

```python
# 옵션 설정 안 하면 species 컬럼 (object 타입) 때문에 오류 남
df.mean(numeric_only=True)
```

```python
df["sepal_length"].mean()
```

#### 10) median() : 중앙값

```python
df.median(numeric_only=True)
```

```python
df["sepal_length"].median()
```

#### 11) quantile()

image.png

```python
df.quantile([0.25, 0.75],numeric_only=True)
```

```python
df['sepal_length'].quantile([0.25, 0.75])
```

#### 12) max()

```python
df.max()
```

#### 13) min()

```python
df.min()
```

#### 14) var() : 분산

```python
df.var(numeric_only=True)
```

#### 15) std() : 표준편차

```python
df.std(numeric_only=True)
```
